```python
class Node:
  def __init__(self, dataval = None):
    self.dataval = dataval
    self.nextval = None

# Membuat node
n1 = Node('Januari')
n2 = Node('Februari')

class LinkedList:
  def __init__(self):
    self.headval = None

  # Menampilkan isi linked list
  def listprint(self):
    printval = self.headval
    while printval is not None:
      print(printval.dataval)
      printval = printval.nextval

  # Fungsi untuk menambahkan node di tengah
  def AddInbetween(self, mid_node, newdata):
    if mid_node is None:
      print('Node yang dimaksud tidak ditemukan')
      return
    NewNode = Node(newdata)
    NewNode.nextval = mid_node.nextval
    mid_node.nextval = NewNode

Li = LinkedList()
Li.headval = n1

# Menghubungkan node pertama ke node kedua
Li.headval.nextval = n2

# Menambahkan node setelah n1
Li.AddInbetween(n1, 'Middle')
Li.listprint()
```

Penyisipan di akhir

```python
class Node:
  def __init__(self, dataval = None):
    self.dataval = dataval
    self.nextval = None

# Membuat node
n1 = Node('Januari')
n2 = Node('Februari')

class LinkedList:
  def __init__(self):
    self.headval = None

  # Menampilkan isi linked list
  def listprint(self):
    printval = self.headval
    while printval is not None:
      print(printval.dataval)
      printval = printval.nextval

  # Fungsi untuk menambahkan node di akhir
  def AddInEnd(self, newdata):
    NewNode = Node(newdata)
    if self.headval is None:
      self.headval = NewNode
      return
    last = self.headval
    while(last.nextval):
      last = last.nextval
    last.nextval = NewNode

Li = LinkedList()
Li.headval = n1

# Menghubungkan node pertama ke node kedua
Li.headval.nextval = n2

# Menambahkan node di bagian akhir
Li.AddInEnd("The Last")
Li.listprint()
```

Hapus node dengan kata kunci

```python
class Node:
  def __init__(self, dataval = Node):
    self.dataval = dataval
    self.nextval = None

class LinkedList:
  def __init__(self):
    self.headval = None

  # Menampilkan isi linked list
  def listprint(self):
    printval = self.headval
    while printval is not None:
      print(printval.dataval)
      printval = printval.nextval

  # Fungsi untuk menambahkan node di awal
  def AddBegining(self, newdata):
    NewNode = Node(newdata)
    # Mengatur nextval node baru agar menunjuk ke head lama
    NewNode.nextval = self.headval
    self.headval = NewNode

  # Fungsi untuk menghapus node berdasarkan kata kunci
  def RemoveNode(self, RemoveKey):
    HeadVal = self.headval

    # Jika node pertama bukan kosong dan datanya sesuai dengan kunci
    if (HeadVal is not None):
      if (HeadVal.dataval == RemoveKey):
        self.headval = HeadVal.nextval
        HeadVal = None
        return

    # Menelusuri node untuk mencari data yang sesuai dengan kunci
    while (HeadVal is not None):
      if HeadVal.dataval == RemoveKey:
        break
      prev = HeadVal
      HeadVal = HeadVal.nextval

    # Jika data tidak ditemukan
    if (HeadVal == None):
      return

    # Menghubungkan node sebelumnya ke node setelah yang dihapus
    prev.nextval = HeadVal.nextval
    HeadVal = None

# Membuat linked list bernama Li
Li = LinkedList()

# Menambahkan beberapa node di awal
Li.AddBegining('April')
Li.AddBegining('Maret')
Li.AddBegining('Februari')

print('Sebelum dilakukan remove')
Li.listprint()
print('==============================================')

# Menghapus node dengan kunci "Maret"
Li.RemoveNode('Maret')

print('Setelah dilakukan remove dengan kunci "Maret"')
Li.listprint()
```

Hapus di akhir

```python
class Node:
  def __init__(self, dataval = Node):
    self.dataval = dataval
    self.nextval = None

class LinkedList:
  def __init__(self):
    self.headval = None

  # Menampilkan isi linked list
  def listprint(self):
    printval = self.headval
    while printval is not None:
      print(printval.dataval)
      printval = printval.nextval

  # Fungsi untuk menambahkan node di akhir
  def AddInEnd(self, newdata):
    NewNode = Node(newdata)
    if self.headval is None:
      self.headval = NewNode
      return
    last = self.headval
    while(last.nextval):
      last = last.nextval
    last.nextval = NewNode

  # Fungsi untuk menghapus node terakhir
  def RemoveEnd(self):
    last = self.headval
    while (last is not None):
      if last.nextval == None:
        break
      prev = last
      last = last.nextval
    if (last == None):
      return
    prev.nextval = last.nextval
    last = None

Li = LinkedList()

# Menambahkan beberapa node
Li.AddInEnd('Januari')
Li.AddInEnd('Februari')
Li.AddInEnd('Maret')

print('Sebelum dilakukan remove')
Li.listprint()
print('==============================================')

# Menghapus node terakhir
Li.RemoveEnd()

print('Setelah dilakukan remove node terakhir')
Li.listprint()
```

```python
class Node:
    def __init__(self, dataval):
        self.dataval = dataval
        self.nextval = None
        self.prevval = None  # Penunjuk ke node sebelumnya


class DoubleLinkedList:
    def __init__(self):
        self.headval = None

    # Menampilkan isi double linked list
    def listprint(self):
        printval = self.headval
        while printval is not None:
            print(printval.dataval)
            printval = printval.nextval

    # Menambahkan node di akhir
    def AddEnd(self, newdata):
        NewNode = Node(newdata)
        if self.headval is None:
            self.headval = NewNode
            return

        last = self.headval
        while last.nextval:
            last = last.nextval

        last.nextval = NewNode
        NewNode.prevval = last

    # Menyisipkan node di tengah (setelah node tertentu)
    def AddAfter(self, prev_node, newdata):
        if prev_node is None:
            print("Node tidak ditemukan")
            return

        NewNode = Node(newdata)
        NewNode.nextval = prev_node.nextval
        prev_node.nextval = NewNode
        NewNode.prevval = prev_node

        if NewNode.nextval is not None:
            NewNode.nextval.prevval = NewNode

    # Menghapus node berdasarkan kata kunci
    def RemoveNode(self, key):
        temp = self.headval

        while temp is not None:
            if temp.dataval == key:
                break
            temp = temp.nextval

        if temp is None:
            print("Data tidak ditemukan")
            return

        if temp.prevval is not None:
            temp.prevval.nextval = temp.nextval
        else:
            self.headval = temp.nextval

        if temp.nextval is not None:
            temp.nextval.prevval = temp.prevval


# Membuat double linked list
DL = DoubleLinkedList()
DL.AddEnd("Januari")
DL.AddEnd("Februari")
DL.AddEnd("Maret")

print("Sebelum penyisipan dan penghapusan:")
DL.listprint()

print("=================================")

# Sisip di tengah (setelah Januari)
DL.AddAfter(DL.headval, "Tengah")

# Hapus dengan kata kunci
DL.RemoveNode("Februari")

print("Setelah penyisipan tengah dan hapus 'Februari':")
DL.listprint()
```

```python
class Node:
    def __init__(self, dataval):
        self.dataval = dataval
        self.nextval = None


class CircularLinkedList:
    def __init__(self):
        self.headval = None

    # Menampilkan isi circular linked list
    def listprint(self):
        temp = self.headval
        if temp is None:
            return

        while True:
            print(temp.dataval)
            temp = temp.nextval
            if temp == self.headval:
                break

    # Menambahkan node di akhir
    def AddEnd(self, newdata):
        NewNode = Node(newdata)

        if self.headval is None:
            self.headval = NewNode
            NewNode.nextval = self.headval
            return

        temp = self.headval
        while temp.nextval != self.headval:
            temp = temp.nextval

        temp.nextval = NewNode
        NewNode.nextval = self.headval

    # Menghapus node terakhir
    def RemoveEnd(self):
        if self.headval is None:
            return

        temp = self.headval
        prev = None

        while temp.nextval != self.headval:
            prev = temp
            temp = temp.nextval

        if prev is None:
            self.headval = None
        else:
            prev.nextval = self.headval


# Membuat circular linked list
CL = CircularLinkedList()
CL.AddEnd("Januari")
CL.AddEnd("Februari")
CL.AddEnd("Maret")

print("Sebelum penyisipan dan penghapusan:")
CL.listprint()

print("=================================")

# Sisip di akhir
CL.AddEnd("Akhir")

# Hapus node terakhir
CL.RemoveEnd()

print("Setelah penyisipan akhir dan hapus akhir:")
CL.listprint()
```
