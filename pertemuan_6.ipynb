Bubble Sort

```python
def bubble_sort(arr):
  n = len(arr)
  for i in range(n-1):
    for j in range(n - i - 1):
      if arr[j] > arr[j + 1]:
        arr[j], arr[j + 1] = arr[j + 1], arr[j]
  return arr

# Contoh penggunaan
data = [64, 34, 25, 12, 22, 11, 90]
sorted_data = bubble_sort(data.copy()) # Menggunakan copy agar data asli tidak berubah
print(f'Data sebelum sorting: {data}')
print(f'Data setelah Bubble Sort: {sorted_data}')
```

Selection Sort

```python
def selection_sort(arr):
  n = len(arr)
  for i in range(n - 1):
    min_index = i
    for j in range(i + 1, n):
      if arr[j] < arr[min_index]:
        min_index = j
    arr[i], arr[min_index] = arr[min_index], arr[i]
  return arr

# Contoh penggunaan
data = [64, 25, 12, 22, 11]
sorted_data = selection_sort(data.copy())
print(f'Data sebelum sorting: {data}')
print(f'Data setelah Selection Sort: {sorted_data}')
```

Insertion Sort

```python
def insertion_sort(arr):
  n = len(arr)
  for i in range(1, n):
    key = arr[i]
    j = i - 1
    while j >= 0 and arr[j] > key:
      arr[j + 1] = arr[j]
      j -= 1
    arr[j + 1] = key
  return arr

# Contoh penggunaan
data = [64, 34, 25, 12, 22, 11, 90]
sorted_data = insertion_sort(data.copy())
print(f'Data sebelum sorting: {data}')
print(f'Data setelah Insertion Sort: {sorted_data}')
```

Merge Sort

```python
def merge_sort(arr):
  if len(arr) > 1:
    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]

    merge_sort(left_half)
    merge_sort(right_half)

    i = j = k = 0

    while i < len(left_half) and j < len(right_half):
      if left_half[i] < right_half[j]:
        arr[k] = left_half[i]
        i += 1
      else:
        arr[k] = right_half[j]
        j += 1
      k += 1

    while i < len(left_half):
      arr[k] = left_half[i]
      i += 1
      k += 1

    while j < len(right_half):
      arr[k] = right_half[j]
      j += 1
      k += 1

  return arr

# Contoh penggunaan
data = [64, 25, 12, 22, 11]
sorted_data = merge_sort(data.copy())
print(f'Data sebelum sorting: {data}')
print(f'Data setelah Merge Sort: {sorted_data}')
```

Quick Sort

```python
def quick_sort(arr):
  if len(arr) <= 1:
    return arr

  pivot = arr[len(arr) // 2]
  left = [x for x in arr if x < pivot]
  middle = [x for x in arr if x == pivot]
  right = [x for x in arr if x > pivot]

  return quick_sort(left) + middle + quick_sort(right)

# Contoh penggunaan
data = [64, 25, 12, 22, 11]
sorted_data = quick_sort(data.copy())
print(f'Data sebelum sorting: {data}')
print(f'Data setelah Quick Sort {sorted_data}')
```

Penugasan

```python
def quick_sort(data):
  if len(data) <= 1:
    return data

  pivot = data[len(data) // 2][1]
  left = [x for x in data if x[1] > pivot]
  middle = [x for x in data if x[1] == pivot]
  right = [x for x in data if x[1] < pivot]

  return quick_sort(left) + middle + quick_sort(right)

# Data mahasiswa (nama, nilai)
mahasiswa = [
  ('Andi', 85),
  ('Budi', 60),
  ('Citra', 70),
  ('Dewi', 90),
  ('Eka', 75),
  ('Fajar', 95),
  ('Gina', 40)
]

sorted_data = quick_sort(mahasiswa)

# Menampilkan peringkat
print('---- Peringkat Mahasiswa ----\n')
for i, (nama, nilai) in enumerate(sorted_data, start = 1):
  print(f'Peringkat {i}: {nama} - Nilai {nilai}')
```
