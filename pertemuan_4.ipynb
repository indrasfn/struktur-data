Queue dengan List

```python
# Program Python untuk mendemonstrasikan implementasi queue (antrian) menggunakan list
queue = []

# fungsi append() digunakan untuk menambahkan elemen ke dalam antrian (enqueue)
queue.append('a')
queue.append('i')
queue.append('u')
queue.append('e')
queue.append('o')

print('Antrian awal:')
print(queue)

# fungsi pop(0) digunakan untuk menghapus elemen dari antrian sesuai urutan FIFO (First In First Out)
print('\nElemen yang dikeluarkan dari antrian')
print(queue.pop(0))
print(queue.pop(0))
print(queue.pop(0))

print('\nAntrian setelah beberapa elemen dihapus')
print(queue)

# jika baris print(queue.pop()) di bawah ini diaktifkan, akan terjadi IndexError karena antrian sudah kosong
```

Queue dengan collections.dequeue

```python
# Program Python untuk mendemonstrasikan implementasi queue menggunakan collections.deque
from collections import deque

queue = deque()

# fungsi append() digunakan untuk menambahkan elemen ke dalam queue (enqueue)
queue.append('a')
queue.append('i')
queue.append('u')
queue.append('e')
queue.append('o')

print('Queue awal:')
print(queue)

# fungsi popleft() digunakan untuk menghapus elemen dari queue sesuai urutan FIFO (First In First Out)
print('\nElemen yang dikeluarkan dari queue')
print(queue.popleft())
print(queue.popleft())
print(queue.popleft())

print('\nQueue setelah beberapa elemen dihapus')
print(queue)

# jika baris print(queue.popleft()) di bawah ini diaktifkan,
# akan terjadi IndexError karena queue sudah kosong
```

Queue dengan queue.Queue

```python
# Program Python untuk mendemonstrasikan implementasi queue menggunakan modul queue
from queue import Queue

# Menginisialisasi sebuah queue
qu = Queue(maxsize=5)

# qsize() menampilkan jumlah elemen yang ada di dalam queue
print(qu.qsize())

# fungsi put() digunakan untuk menambahkan elemen ke dalam queue (enqueue)
qu.put('a')
qu.put('i')
qu.put('u')
qu.put('e')
qu.put('o')

print(f'Queue penuh: {qu.full()}')
print(f'Queue kosong: {qu.empty()}')

# fungsi get() digunakan untuk mengambil elemen dari queue (dequeue)
print('\nElemen yang dikeluarkan dari queue')
print(qu.get())
print(qu.get())
print(qu.get())

print(f'\nQueue kosong: {qu.empty()}')
print(f'Queue penuh: {qu.full()}')
```

Queue dengan Single Linked List

```python
class Queue:
  def __init__(self):
    self.queue = list()

  def addtoqu(self, dataval):
    if dataval not in self.queue:
      self.queue.insert(0, dataval)
      return True
    return False

  def removerfromq(self):
    if len(self.queue) > 0:
      return self.queue.pop()
    return ('No elements in Queue')

  def size(self):
    return len(self.queue)
TheQueue = Queue()
print(TheQueue.addtoqu('Jan'))
print(TheQueue.addtoqu('Feb'))
print(TheQueue.addtoqu('March'))
print(TheQueue.addtoqu('April'))
print(TheQueue.size())
print(TheQueue.removerfromq())
print(TheQueue.removerfromq())
print(TheQueue.size())
```
