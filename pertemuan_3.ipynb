Stack dengan List

```python
# Program Python untuk mendemonstrasikan implementasi stack menggunakan list

stack = []

# Fungsi append() digunakan untuk menambahkan elemen ke dalam stack (push)
stack.append('a')
stack.append('i')
stack.append('u')
stack.append('e')
stack.append('o')

print('Stack awal')
print(stack)

# Fungsi pop() digunakan untuk mengambil elemen dari stack sesuai prinsip LIFO
print('\nElemen yang diambil dari stack:')
print(stack.pop())
print(stack.pop())
print(stack.pop())

print('\nStack setelah beberapa elemen diambil')
print(stack)

# Jika print(stack.pop()) dibuka terus sampai kosong,
# maka akan menyebabkan IndexError karena stack sudah kosong
```

Stack dengan collections.deque

```python
# Program Python untuk mendemonstrasikan implementasi stack menggunakan collections.deque

from collections import deque

stack = deque()

# Fungsi append() digunakan untuk menambahkan elemen ke dalam stack (push)
stack.append('a')
stack.append('i')
stack.append('u')
stack.append('e')
stack.append('o')

print('Stack awal')
print(stack)

# Fungsi pop() digunakan untuk mengambil elemen dari stack sesuai prinsip LIFO
print('\nElemen yang diambil dari stack')
print(stack.pop())
print(stack.pop())
print(stack.pop())

print('\nStack setelah beberapa elemen diambil:')
print(stack)

# Jika print(stack.pop()) dibuka terus sampai kosong,
# maka akan menyebabkan IndexError karena stack sudah kosong
```

Stack dengan queue.LifoQueue

```python
# Program Python untuk mendemonstrasikan implementasi stack menggunakan module queue

from queue import LifoQueue

# Membuat stack dengan kapasitas maksimal 5 elemen
stack = LifoQueue(maxsize=5)

# qsize() menampilkan jumlah elemen dalam stack
print(stack.qsize())

# put() digunakan untuk menambahkan elemen ke dalam stack (push)
stack.put('a')
stack.put('i')
stack.put('u')
stack.put('e')
stack.put('o')

print(f'Penuh: {stack.full()}')
print(f'Ukuran: {stack.qsize()}')

# get() digunakan untuk mengambil elemen dari stack sesuai prinsip LIFO
print('\nElemen yang diambil dari stack')
print(stack.get())
print(stack.get())
print(stack.get())

print(f'\nKosong: {stack.empty()}')
```

Stack dengan Single Linked List

```python
class Node:
  def __init__(self, value):
    self.value = value
    self.next = None

class Stack:
  # Menginisialisasi stack, menggunakan dummy node agar lebih mudah menangani kondisi khusus
  def __init__(self):
    self.head = Node('head')
    self.size = 0

  # Representasi string dari stack
  def __str__(self):
    cur = self.head.next
    out = ''
    while cur:
      out += str(cur.value) + '->'
      cur = cur.next
    return out[:-2]

  # Mengambil ukuran stack saat ini
  def getSize(self):
    return self.size

  # Mengecek apakah stack kosong
  def isEmpty(self):
    return self.size == 0

  # Mengambil elemen paling atas dari stack
  def peek(self):
    # Pengecekan untuk memastikan tidak mengambil dari stack kosong
    if self.isEmpty():
      raise Exception('Melihat dari stack kosong')
    return self.head.next.value

  # Menambahkan nilai ke dalam stack
  def push(self, value):
    node = Node(value)
    node.next = self.head.next
    self.head.next = node
    self.size += 1

  # Menghapus nilai dari stack dan mengembalikannya
  def pop(self):
    if self.isEmpty():
      raise Exception('Menghapus dari stack kosong')
    remove = self.head.next
    self.head.next = self.head.next.next
    self.size -= 1
    return remove.value

# Kode Utama (Driver Code)
if __name__ == '__main__':
  stack = Stack()
  for i in range(1, 11):
    stack.push(i)
  print(f'Stack: {stack}')

  for _ in range(1, 6):
    remove = stack.pop()
    print(f'Pop: {remove}')
  print(f'Stack: {stack}')
```

Queue menggunakan collections.deque

```python
from collections import deque

queue = deque()

# enqueue (menambahkan data ke antrian)
queue.append("Dokumen A")
queue.append("Dokumen B")
queue.append("Dokumen C")

print("Antrian cetak:", queue)

# dequeue (mengambil data dari antrian)
print("Mencetak:", queue.popleft())
print("Mencetak:", queue.popleft())

print("Sisa antrian:", queue)
```
