Graf dengan Adjacency List

```python
# Membuat graf tak berarah menggunakan adjacency list

graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

# Menampilkan graf
def print_graph(graf):
  for node in graf:
    print(f'{node} -> {graf[node]}')

print_graph(graph)
```

Traversal Graf menggunakan BFS

```python
from collections import deque

def bfs(graph, start):
  visited = set()
  queue = deque([start])

  while queue:
    vertex = queue.popleft()
    if vertex not in visited:
      print(vertex, end=' ')
      visited.add(vertex)
      queue.extend([neighbor for neighbor in graph[vertex] if neighbor not in visited])

# Contoh pemanggilan
bfs(graph, 'A')
```

Traversal Graf menggunakan DFS

```python
def dfs(graph, start, visited = None):
  if visited is None:
    visited = set()

  if start not in visited:
    print(start, end=' ')
    visited.add(start)
    for neighbor in graph[start]:
      dfs(graph, neighbor, visited)

# Contoh pemanggilan
dfs(graph, 'A')
```

Latihan Personal

```python
def count_neighbors(graph):
    result = {}
    for node in graph:
        result[node] = len(graph[node])
    return result

print(count_neighbors(graph))
```

```python
from collections import deque

# BFS
def bfs(graph, start):
    visited = set()
    queue = deque([start])
    print(f"BFS dari '{start}': ", end="")

    while queue:
        vertex = queue.popleft()
        if vertex not in visited:
            print(vertex, end=" ")
            visited.add(vertex)
            queue.extend([n for n in graph[vertex] if n not in visited])
    print()

# DFS
def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()
        print(f"DFS dari '{start}': ", end="")

    if start not in visited:
        print(start, end=" ")
        visited.add(start)
        for neighbor in graph[start]:
            dfs(graph, neighbor, visited)

    if len(visited) == len(graph):
        print()

# Jalankan keduanya
bfs(graph, 'Jakarta')
dfs(graph, 'Jakarta')
```

```python
def find_path(graph, start, end, visited=None, path=None):
    if visited is None:
        visited = set()
    if path is None:
        path = []

    visited.add(start)
    path = path + [start]

    # Kalau sudah sampai tujuan, kembalikan path-nya
    if start == end:
        return path

    # Telusuri semua tetangga
    for neighbor in graph[start]:
        if neighbor not in visited:
            result = find_path(graph, neighbor, end, visited, path)
            if result:
                return result
    return None

# Contoh penggunaan
jalur1 = find_path(graph, 'Jakarta', 'Bandung')
print(f"Jakarta ke Bandung: {jalur1}")

jalur2 = find_path(graph, 'Bekasi', 'Bogor')
print(f"Bekasi ke Bogor: {jalur2}")

# Coba dari simpul ke dirinya sendiri
jalur3 = find_path(graph, 'Depok', 'Depok')
print(f"Depok ke Depok: {jalur3}")
```

```python
def is_connected(graph):
    if not graph:
        return True

    start = next(iter(graph))
    visited = set()

    # DFS helper (versi iteratif dengan stack)
    def dfs_helper(node):
        visited.add(node)
        for neighbor in graph[node]:
            if neighbor not in visited:
                dfs_helper(neighbor)

    dfs_helper(start)

    # Kalau semua simpul terkunjungi, berarti graf terhubung
    return len(visited) == len(graph)

# Pengujian 1: Graf yang terhubung
print(f"Graf kota terhubung? {is_connected(graph)}")

# Pengujian 2: Graf yang TIDAK terhubung
graph_putus = {
    'A': ['B'],
    'B': ['A'],
    'C': ['D'],  # C dan D tidak terhubung ke A dan B
    'D': ['C']
}
print(f"Graf putus terhubung? {is_connected(graph_putus)}")
```
